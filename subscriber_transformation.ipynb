```python
sub_df = sub_df.withColumn("Subgrp_id", when((col("Subgrp_id").isNull()) & (col("sub_id") == "SUBID10022"), "S110") \
                                       .when((col("Subgrp_id").isNull()) & (col("sub_id") == "SUBID10049"), "S107") \
                                        .otherwise(col("Subgrp_id")))
```
